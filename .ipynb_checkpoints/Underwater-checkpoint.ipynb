```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import tkinter
```

### CV2 Version

```python
img_bgr=cv.imread("coral.jpg", cv.IMREAD_COLOR)

#Convert image from BGR to LAB
img_lab=cv.cvtColor(img_bgr, cv.COLOR_BGR2LAB)

#Split channels
l, a, b = cv.split(img_lab)

#Plot alpha vs beta
plt.scatter(a,b, s=8)
plt.show()

#Compute medians for each channel
l_med=np.median(l).astype("int")
a_med=np.median(a).astype("int")
b_med=np.median(b).astype("int")

#Subtraction of median
#a=a-a_med
#b=b-b_med

#Merge lab channels
balanced_lab=cv.merge([l, a, b])

#Convert from LAB to BGR
img_out=cv.cvtColor(balanced_lab, cv.COLOR_LAB2BGR)

#Show output in RGB
plt.imshow(cv.cvtColor(img_out, cv.COLOR_BGR2RGB))
plt.show()

```

### Our version

```python
def RGB2LAB(img_rgb):

    img_rgb=(img_rgb/255)+0.000000001

    T_xyz=np.array([[0.5141, 0.3239, 0.1694],
                    [0.2651, 0.6702, 0.0641],
                    [0.0241, 0.1228, 0.8444]])

    T_lms=np.array([[0.3897, 0.6890,0.0787],
                    [-0.2298, 1.1834, 0.0464],
                    [0,0,1]])

    T_pca_1=np.array([[1/np.sqrt(3),0,0],[0,1/np.sqrt(6),0],[0,0,np.sqrt(2)]])
    T_pca_2=np.array([[1,1,1],[1,1,-2],[1,-1,0]])

    T_pca=np.dot(T_pca_1,T_pca_2)
    
    img_lms=np.matmul(np.matmul(img_rgb,T_xyz.T),T_lms.T)

    img_lab=np.matmul(np.log10(img_lms),T_pca.T)
    
    return img_lab


def LAB2RGB(img_lab):
    
    T_xyz_inv=np.linalg.inv(np.array([[0.5141, 0.3239, 0.1694],
                    [0.2651, 0.6702, 0.0641],
                    [0.0241, 0.1228, 0.8444]]))

    T_lms_inv=np.linalg.inv(np.array([[0.3897, 0.6890,0.0787],
                    [-0.2298, 1.1834, 0.0464],
                    [0,0,1]]))

    T_pca_1=np.array([[1/np.sqrt(3),0,0],[0,1/np.sqrt(6),0],[0,0,np.sqrt(2)]])
    T_pca_2=np.array([[1,1,1],[1,1,-2],[1,-1,0]])
    
    T_pca=np.dot(T_pca_1,T_pca_2)
    
    T_pca_inv=np.linalg.inv(T_pca)
    
    img_lms=10**(np.matmul(img_lab,T_pca_inv.T))

    img_rgb=np.matmul(np.matmul(img_lms,T_lms_inv.T),T_xyz_inv.T)
    
    img_rgb=img_rgb
    
    return img_rgb
```

```python
img_bgr=cv.imread("coral.jpg", cv.IMREAD_COLOR)
img_rgb=cv.cvtColor(img_bgr, cv.COLOR_BGR2RGB)

img_lab=RGB2LAB(img_rgb)

img_rgb_out=LAB2RGB(img_lab)

plt.imshow(img_rgb)
plt.show()
plt.imshow(img_rgb_out)
plt.show()

print(img_rgb_out.max(), img_rgb_out.min())
```

```python
img_bgr=cv.imread("coral.jpg", cv.IMREAD_COLOR)
img_rgb=cv.cvtColor(img_bgr, cv.COLOR_BGR2RGB)

img_lab=RGB2LAB(img_rgb)

l, a, b = cv.split(img_lab)

a_med=np.median(a)
b_med=np.median(b)

print("Alpha channel min, median, max:", a.min(), a_med, a.max())
print("Beta channel min, median, max:", b.min(), b_med, b.max())

a_balanced=a-a_med
b_balanced=b-b_med

img_lab_balanced=cv.merge([l, a_balanced, b_balanced])

img_rgb_balanced=LAB2RGB(img_lab_balanced)

plt.imshow(img_rgb)
plt.show()
plt.imshow(img_rgb_balanced)
plt.show()
```
